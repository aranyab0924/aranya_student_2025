```python
public class Scoreboard {
    private String team1Name;
    private String team2Name;
    private int team1Score;
    private int team2Score;
    private boolean isTeam1Active;

    // The parameters here (t1, t2) must match the names used below
    public Scoreboard(String t1, String t2) {
        team1Name = t1;
        team2Name = t2;
        team1Score = 0;
        team2Score = 0;
        isTeam1Active = true;
    }

    public void recordPlay(int points) {
        if (points > 0) {
            if (isTeam1Active) {
                team1Score += points;
            } else {
                team2Score += points;
            }
        } else {
            // Points is 0, so switch the active team
            isTeam1Active = !isTeam1Active;
        }
    }

    public String getScore() {
        String activeName;
        if (isTeam1Active) {
            activeName = team1Name;
        } else {
            activeName = team2Name;
        }
        
        // Final check: Note the \"+\" before activeName
        return team1Score + \"-\" + team2Score + \"-\" + activeName;
    }
}

// Driver to reproduce the example output from the original pages runner
class Main {
    public static void main(String[] args) {
        Scoreboard s = new Scoreboard("Red", "Blue");
        // Step numbers are for reference to match the example trace
        // Step 1 - initial
        System.out.println("(Step 1) info = " + s.getScore());
        // Step 2 - toggle (points = 0)
        s.recordPlay(0);
        System.out.println("(Step 2) info = " + s.getScore());
        // Step 3 - Red scores 1
        s.recordPlay(1);
        System.out.println("(Step 3) info = " + s.getScore());
        // Step 4 - toggle
        s.recordPlay(0);
        System.out.println("(Step 4) info = " + s.getScore());
        // Step 5 - Blue scores 3
        s.recordPlay(3);
        System.out.println("(Step 5) info = " + s.getScore());
        // Additional steps to illustrate behavior
        s.recordPlay(0); // switch
        s.recordPlay(1); // Red +1
        s.recordPlay(7); // Red +7
        System.out.println("(Final) info = " + s.getScore());
    }
}
```
