```python
public class Scoreboard { private String team1Name; private String team2Name; private int team1Score; private int team2Score; private boolean isTeam1Active;

// The parameters here (t1, t2) must match the names used below
public Scoreboard(String t1, String t2) {
    team1Name = t1;
    team2Name = t2;
    team1Score = 0;
    team2Score = 0;
    isTeam1Active = true;
}

public void recordPlay(int points) {
    if (points > 0) {
        if (isTeam1Active) {
            team1Score += points;
        } else {
            team2Score += points;
        }
    } else {
        // Points is 0, so switch the active team
        isTeam1Active = !isTeam1Active;
    }
}

public String getScore() {
    String activeName;
    if (isTeam1Active) {
        activeName = team1Name;
    } else {
        activeName = team2Name;
    }
    
    // Final check: Note the "+" before activeName
    return team1Score + "-" + team2Score + "-" + activeName;
} }
```

Challenges

- Keeping track of which team is active and ensuring the toggle (when points == 0) happens at the correct time are common pitfalls.
- Off-by-one or incorrect accumulation errors can happen if the active flag isn't consistently updated.

What I learned

- Use small driver tests to verify state transitions (score updates and active-team toggles).
- Keep method responsibilities small: one method to update state (recordPlay) and one to format output (getScore).
